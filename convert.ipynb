```python
def back(x):
    #print(*reversed(tuple(range(x.dim()))))
    return x.permute(*reversed(tuple(range(x.dim()))))
def back_conv(x):
    #print(*reversed(tuple(range(x.dim()))))
    return x.permute(3,2,0,1)
```
